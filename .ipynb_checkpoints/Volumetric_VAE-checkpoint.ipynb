```python
def pcd2voxel(pcd, filter_rad=0.1, dim_size=32):
    [xc, yc, zc] = np.mean(pcd.points,0)
    [xs, ys, zs] = [xc, yc, zc] + np.ones(3)*filter_rad*dim_size/2
    [xe, ye, ze] = [xc, yc, zc] - np.ones(3)*filter_rad*dim_size/2

    voxel_volume = np.zeros((dim_size,dim_size,dim_size))
    color_volume = np.zeros((dim_size,dim_size,dim_size,3))
    for point, color in zip(pcd.points, pcd.colors):
        xi,yi,zi = (np.around(([xs, ys, zs] - point)/filter_rad)).astype(int)
        voxel_volume[xi, yi, zi] = 1
        color_volume[xi, yi, zi]= color
    
    return voxel_volume, color_volume

def vis_volume(pcd):
    voxel_volume, color_volume = pcd2voxel(pcd)
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')
    ax.voxels(voxel_volume, facecolors=color_volume)
    plt.show()
    
def vis_numpy_volume(numpy_volume):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')
    ax.voxels(numpy_volume)
    plt.show()
```

```python
class Encoder(torch.nn.Module):
    def __init__(self, _kernel_size=2, _stride=1):
        super(Encoder, self).__init__()
        self.conv3d_1 = torch.nn.Conv3d(1,20,1,1)
        self.conv3d_2 = torch.nn.Conv3d(20,128,_kernel_size,_stride,padding=1)
        self.pool3d_2 = torch.nn.MaxPool3d(2)
        self.conv3d_3 = torch.nn.Conv3d(128,64,_kernel_size,_stride,padding=1)
        self.pool3d_3 = torch.nn.MaxPool3d(2)
        self.conv3d_4 = torch.nn.Conv3d(64,32,_kernel_size,_stride,padding=1)
        self.pool3d_4 = torch.nn.MaxPool3d(2)

    def forward(self, x):
        x = F.relu(self.conv3d_1(x))
        x = F.relu(self.conv3d_2(x))
        x = self.pool3d_2(x)
        x = F.relu(self.conv3d_3(x))
        x = self.pool3d_3(x)
        x = F.relu(self.conv3d_4(x))
        x = self.pool3d_4(x)
        x = x.view(-1)
        
        return x

class Decoder(torch.nn.Module):
    def __init__(self, _kernel_size=2,_stride=1):
        super(Decoder, self).__init__()
        self.conv3d_1 = torch.nn.Conv3d(32,64,1,1)
        self.upsample_1 = torch.nn.Upsample(scale_factor=2, mode="nearest")
        self.conv3d_2 = torch.nn.Conv3d(64,128,1,1)
        self.upsample_2 = torch.nn.Upsample(scale_factor=2, mode="nearest")
        self.conv3d_3 = torch.nn.Conv3d(128,20,1,1)
        self.upsample_3 = torch.nn.Upsample(scale_factor=2, mode="nearest")
        self.softmax = torch.nn.Softmax()

    def forward(self, x):
        x = self.conv3d_1(x)
        x = F.relu(self.upsample_1(x))
        x = self.conv3d_2(x)
        x = F.relu(self.upsample_2(x))
        x = self.conv3d_3(x)
        x = F.relu(self.upsample_3(x))
        x = self.softmax(x)
        
        return x
    

    
class Volumetric_VAE(torch.nn.Module):    
    def __init__(self,_kernel_size=2,_stride=1):
        super(Volumetric_VAE, self).__init__()
        self.encoder = Encoder(_kernel_size, _stride)
        self.decoder = Decoder(_kernel_size, _stride)
        
        self._mu = torch.nn.Linear(2048, 256)
        self._log_sigma = torch.nn.Linear(2048,256)
        
        self._z = torch.nn.Linear(256,2048)
        
    def _sample_latent(self, bottle_neck):
        mu = F.tanh(self._mu(bottle_neck))
        log_sigma = F.tanh(self._log_sigma(bottle_neck))
        
        sigma=torch.exp(log_sigma)
        std_z = torch.from_numpy(np.random.normal(0,1,size=sigma.size())).float()
        
        self.z_mean = mu
        self.z_sigma = sigma
        
        return mu + sigma * Variable(std_z, requires_grad=False) # Reparameterization trick
    
    def forward(self, x):
        bottle_neck = self.encoder.forward(x)
        z = self._sample_latent(bottle_neck)
        z = self._z(z)
        z = z.view([-1,32,4,4,4])
        output = self.decoder.forward(z)
        
        return output
        
```

```python
def is_in_palette(color_palette, c):
    flag = False
    for color in color_palette:
        if(np.array_equal(color, c)):
            flag = True
    
    return flag

pcd = read_point_cloud("/home/dongwonshin/Desktop/subvolume_dataset_making/mesh_results2/000042_complete.ply")
voxel_volume, color_volume = pcd2voxel(pcd)
vis_volume(pcd)

color_values = color_volume.reshape([-1,3])
color_palette = []
for color_value in color_values:
    if (not np.array_equal(color_value, [0,0,0])):
        if (not is_in_palette(color_palette, color_value)):
            color_palette.append(color_value)

palette =  np.asarray(color_palette)
indices = np.asarray([range(len(color_palette))])
io.imshow(palette[indices])     
```

```python
nyudv2_color_codes = { 
    'floor': [162, 227, 148], 
    'wall': [183, 205, 235], 
    'cabinet': [35, 131, 188], 
    'bed': [255, 196, 131], 
    'chair': [195, 197, 37], 
    'sofa': [151, 97, 84], 
    'table': [255, 162, 161], 
    'door': [218, 45, 45], 
    'window': [203, 185, 218], 
    'bookshelf': [159, 114, 196], 
    'picture': [203, 166, 158], 
    'counter': [24, 197, 214], 
    'desk': [248, 191, 216], 
    'curtain': [222, 224, 152], 
    'refrigerator': [254, 139, 10], 
    'bathtub':[231, 130, 201], 
    'shower_curtain': [168, 223, 232], 
    'toilet': [48, 170, 51], 
    'sink':[123, 139, 155], 
    'otherfuniture':[93, 93, 173] 
}
    
def rgb2nyudv2(rgb):
    
    color_dist = []
    for class_name, code in nyudv2_color_codes.items():
        color_dist.append(np.linalg.norm(np.asarray(code)-rgb))
    
    class_id = np.argmin(color_dist)
    class_name = list(nyudv2_color_codes.keys())[class_id]
    
    return class_id, class_name
```

```python
def colorvol2classidvol(color_volume):
    classid_volume = np.zeros([32,32,32])
    classid_volume = classid_volume.flatten()
    color_volume = color_volume.reshape([-1,3])

    for idx, color_voxel in enumerate(color_volume):
        if(not np.array_equal(color_voxel, [0,0,0])):
            class_id, calss_name = rgb2nyudv2(color_voxel*255)
            classid_volume[idx] = class_id
    #         print(class_id,calss_name)

    classid_volume = classid_volume.reshape([32,32,32])
    
    return classid_volume
```

```python
def classidvol2colorvol(classid_volume):
    color_volume = np.zeros([32,32,32,3])
    color_volume = color_volume.reshape([-1,3])
    classid_volume = classid_volume.flatten()
    
    for idx, classid_voxel in enumerate(classid_volume):
        if (classid_voxel != 0):
            color_volume[idx] = list(nyudv2_color_codes.values())[int(classid_voxel)]
    
    color_volume = color_volume.reshape([32,32,32,3])
    
    return color_volume

color_volume = classidvol2colorvol(classidvol)
```
